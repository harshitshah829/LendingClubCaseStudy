```python
# Importing necessary libraries for data analysis
import numpy as np
import pandas as pd

# Importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')  # Suppress warnings
```

```python
# Loading the data from CSV file into a DataFrame
loan_data = pd.read_csv('loan.csv')

# Checking the dimensions of the DataFrame
loan_data.shape
```

```python
# Displaying summary statistics of the data
loan_data.describe()  # Summary of numeric columns
```

```python

# Displaying summary statistics of categorical columns
loan_data.describe(include='O')
```

```python
# Viewing the first few rows of the DataFrame
loan_data.head()
```

```python
# Checking unique values in the 'loan_status' column
loan_data.loan_status.value_counts()
```

```python
# Filtering out records where the loan status is 'Current'
loan_data = loan_data[loan_data['loan_status'] != 'Current']
loan_data.head()
```

```python
# Dropping columns with all null values
loan_data = loan_data.dropna(axis=1, how='all')
loan_data.head()
```

```python
# Dropping columns that do not impact loan default prediction
irrelevant_columns = [
    'desc', 'mths_since_last_delinq', 'mths_since_last_record', 'delinq_2yrs',
    'earliest_cr_line', 'last_pymnt_amnt', 'inq_last_6mths', 'open_acc',
    'pub_rec', 'revol_bal', 'revol_util', 'total_acc', 'total_pymnt', 
    'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee',
    'recoveries', 'collection_recovery_fee', 'last_pymnt_d', 'last_credit_pull_d',
    'title', 'emp_title', 'member_id', 'url', 'tax_liens', 'acc_now_delinq',
    'chargeoff_within_12_mths', 'delinq_amnt', 'initial_list_status', 'out_prncp',
    'out_prncp_inv', 'collections_12_mths_ex_med', 'policy_code', 'application_type',
    'pymnt_plan', 'zip_code'
]
```

```python
loan_data = loan_data.drop(irrelevant_columns, axis=1)
loan_data.shape
```

```python
# Displaying remaining columns after cleaning
loan_data.columns
```

```python
# Removing the 'months' string from the 'term' column
loan_data.term = loan_data.term.str.replace("months", "").astype(int)
```

```python
# Converting interest rate from string to numeric
loan_data.int_rate = loan_data.int_rate.str.replace("%", "").astype(float)
loan_data.dtypes
```

```python
# Converting 'issue_d' from object to datetime
loan_data['issue_d'] = pd.to_datetime(loan_data.issue_d, format='%b-%y')
loan_data['issue_d']
```

```python
# Checking the proportion of missing values
loan_data.isnull().mean()
```

```python
# Adding a new column 'issue_year' to the DataFrame
loan_data["issue_year"] = loan_data["issue_d"].dt.year
loan_data.head()
```

```python
# Displaying summary statistics again after data cleaning
loan_data.describe()
```

```python
# Univariate Analysis

# Distribution plots for Annual Income, Loan Amount, and Interest Rate
features = ["annual_inc", "loan_amnt", "int_rate"]

for feature in features:
    plt.figure(figsize=(8, 4))
    sns.histplot(loan_data[feature], kde=True, color='blue')
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Count plot for the 'purpose' column
plt.figure(figsize=(10, 5))
sns.countplot(y='purpose', data=loan_data, order=loan_data['purpose'].value_counts().index, palette='viridis')
plt.title('Loan Purpose Distribution')
plt.show()
```

```python
# Count plot for the 'addr_state' column
plt.figure(figsize=(15, 6))
sns.countplot(y='addr_state', data=loan_data, order=loan_data['addr_state'].value_counts().index, palette='coolwarm')
plt.title('Loan Distribution by State')
plt.show()
```

```python
# Count plot for the 'emp_length' column
plt.figure(figsize=(10, 5))
sns.countplot(y='emp_length', data=loan_data, order=loan_data['emp_length'].value_counts().index, palette='magma')
plt.title('Employment Length Distribution')
plt.show()
```

```python
# Verification status impact on loan amount using box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='verification_status', y='loan_amnt', data=loan_data, palette='Set2')
plt.title('Loan Amount by Verification Status')
plt.show()
```

```python
# Loan amount distribution by term and grade using box plots
plt.figure(figsize=(16, 6))

sns.boxplot(x='term', y='loan_amnt', data=loan_data, palette='Set1')
plt.title('Loan Amount by Term')

plt.tight_layout()
plt.show()
```

```python
# Loan amount distribution by purpose using box plot
plt.figure(figsize=(14, 6))
sns.boxplot(x='purpose', y='loan_amnt', data=loan_data, palette='pastel')
plt.title('Loan Amount by Purpose')
plt.xticks(rotation=90)
plt.show()
```

```python

# Loan amount distribution by issue year using violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='issue_year', y='loan_amnt', data=loan_data, palette='coolwarm')
plt.title('Loan Amount by Issue Year')
plt.show()
```

```python
# Analysis of loan status proportions by term
term_status_df = loan_data.groupby(['term', 'loan_status']).size().reset_index(name='counts')
term_status_df['proportion'] = term_status_df.groupby('term')['counts'].transform(lambda x: x / x.sum())

plt.figure(figsize=(10, 6))
sns.barplot(x='term', y='proportion', hue='loan_status', data=term_status_df, palette='coolwarm')
plt.title('Proportion of Loan Status by Term')
plt.show()
```

```python
# Box plot and bar plot for loan amount distribution by term and loan status
plt.figure(figsize=(18, 8))

sns.boxplot(x='term', y='loan_amnt', hue='loan_status', data=loan_data, palette='Set1')
plt.title('Loan Amount Distribution by Term and Loan Status')

plt.tight_layout()
plt.show()
```

```python
# Box plot and point plot for interest rate by grade and loan status
plt.figure(figsize=(18, 8))

sns.boxplot(x='grade', y='int_rate', hue='loan_status', data=loan_data, palette='Set1')
plt.title('Interest Rate Distribution by Grade and Loan Status')

plt.tight_layout()
plt.show()
```

```python
# Filtering data for Charged Off loans
charged_off_data = loan_data[loan_data['loan_status'] == 'Charged Off']

# Count plot for Charged Off loans by state
plt.figure(figsize=(14, 6))
sns.countplot(y='addr_state', data=charged_off_data, order=charged_off_data['addr_state'].value_counts().index, palette='Reds')
plt.title('Charged Off Loans by State')
plt.show()
```

```python
# Calculating the percentage of Charged Off loans by purpose
purpose_status_df = loan_data.groupby(['purpose', 'loan_status']).size().reset_index(name='counts')
purpose_status_df['total'] = purpose_status_df.groupby('purpose')['counts'].transform('sum')
purpose_status_df['percentage'] = purpose_status_df.apply(lambda row: row['counts'] / row['total'] if row['loan_status'] == 'Charged Off' else 0, axis=1)

# Filtering the data for Charged Off loans
charged_off_purpose = purpose_status_df[purpose_status_df['loan_status'] == 'Charged Off']

# Pie chart for Charged Off loans by purpose
plt.figure(figsize=(10, 8))
plt.pie(charged_off_purpose['percentage'], labels=charged_off_purpose['purpose'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Percentage of Charged Off Loans by Purpose')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Hexbin plot for loan amount vs. DTI
plt.figure(figsize=(12, 8))

# Plot for Charged Off loans
plt.hexbin(x=charged_off_data['loan_amnt'], y=charged_off_data['dti'], gridsize=50, cmap='Reds', mincnt=1, alpha=0.5, label='Charged Off')

# Plot for Fully Paid loans
plt.hexbin(x=loan_data[loan_data['loan_status'] == 'Fully Paid']['loan_amnt'], y=loan_data[loan_data['loan_status'] == 'Fully Paid']['dti'], gridsize=50, cmap='Greens', mincnt=1, alpha=0.5, label='Fully Paid')

# Adding colorbar
cb = plt.colorbar(label='Count in bin')

plt.title('Loan Amount vs DTI')
plt.xlabel('Loan Amount')
plt.ylabel('DTI')
plt.legend(['Charged Off', 'Fully Paid'])
plt.show()
```

```python
# Correlation analysis - New
corr_columns = ['loan_amnt', 'funded_amnt', 'int_rate', 'annual_inc', 'dti']
corr_matrix = loan_data[corr_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
